```python
sql = '''
WITH data AS(
SELECT code, point, date FROM income
UNION
SELECT code,point,date FROM outcome)

SELECT data.point, data.date, sum(out), sum(inc) FROM data 

LEFT JOIN income ON income.point = data.point and income.date = data.date and income.code = data.code
LEFT JOIN outcome ON outcome.point = data.point and outcome.date = data.date and outcome.code = data.code

GROUP BY data.point,data.date
;
'''
select(sql)
```

### Задание: 39
Найдите корабли, `сохранившиеся для будущих сражений`; т.е. выведенные из строя в одной битве (damaged), они участвовали в другой, произошедшей позже.

```python
sql = '''
WITH data AS(
SELECT ship,result,date FROM Outcomes o
LEFT JOIN Battles b ON b.name = o.battle
WHERE ship IN(SELECT ship FROM outcomes WHERE result = 'damaged')
)

SELECT DISTINCT data.ship  FROM data
LEFT JOIN (SELECT ship,
            min(date) as dmg_date 
                FROM data
                WHERE result = 'damaged' GROUP BY ship) AS query 
ON query.ship = data.ship
WHERE date > dmg_date
;
'''
select(sql)
```

### Задание: 41
Для каждого производителя, у которого присутствуют модели хотя бы в одной из таблиц PC, Laptop или Printer,
определить максимальную цену на его продукцию.

Вывод: имя производителя, если среди цен на продукцию данного производителя присутствует NULL, то выводить для этого производителя NULL,
иначе максимальную цену.

```python
sql = '''
WITH data as(
SELECT model,price FROM PC
UNION
SELECT model,price FROM Laptop
UNION
SELECT model,price FROM Printer
)

SELECT maker,
CASE WHEN MAX(CASE WHEN price IS NULL THEN 1 ELSE 0 END) = 0 
THEN MAX(price) END AS maximim
FROM data 
LEFT JOIN Product ON Product.model = data.model
GROUP BY maker

;
'''
select(sql)
```
